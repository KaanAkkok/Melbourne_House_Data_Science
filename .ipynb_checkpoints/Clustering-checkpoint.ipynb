```python
# datasetin içinden bazı columnları atıcam
# address, sellerG, Date, propertycount, postcode, method bu sütunları attım çünkü gerekli değil
# suburb, rooms, type, bedroom2, bathroom, car, landsize, buildingarea, yearbuild, lattitude, longtitude, regionname atmadım
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from scipy.cluster.hierarchy import dendrogram, linkage
import matplotlib.pyplot as plt
```

```python
sample_df = pd.read_csv("MH_precrocessed_dataset.csv")
```

```python
sample_df_dropped = sample_df.drop(columns=['Address', 'SellerG', 'Date', 'Propertycount', 'Postcode', 'Method'])
```

```python
sample_df_dropped.head()
```

```python
numeric_cols = sample_df_dropped.select_dtypes(include=[np.number]).columns.tolist()
categorical_cols = sample_df_dropped.select_dtypes(include=['object']).columns.tolist() 

print(numeric_cols)
print(categorical_cols)
```

```python
# kategorik değerleri one hot encoding ile numerik değere çevirdim
sample_df_encoded = pd.get_dummies(sample_df_dropped[categorical_cols], drop_first=True)
# numerik ile one hot encoding kategorik değerleri birleştirdim
X = sample_df_dropped[numeric_cols].join(sample_df_encoded)
X = X.drop(columns=['Price']) # price çıkartıyorum çünkü bu bizim targetimiz (class label)
X.head()
```

```python
# standarscaler ile standardlaştırma (bazı değerler yüksek olabilir bundan dolayı hatalı olabilir)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# hiyerarşik clustering için linkage oluşturuyorum (ward - euclid distance)
linked = linkage(X_scaled, method='ward')
```

```python
# dendogram çizimi
plt.figure(figsize=(20,10))
dendrogram(linked, truncate_mode='level', p=5)
plt.title("Dendogram")
plt.xlabel("Data points")
plt.ylabel("Euclidean distance")
plt.tight_layout()
plt.show()
```
