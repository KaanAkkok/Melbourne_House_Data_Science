```python
sample_df = dataset.sample(n=2000, random_state=5006).reset_index(drop=True)
sample_df.head(5)
```

```python
for col in sample_df.columns:
    print(f"{col}: \n\tdtype:{sample_df[col].dtype} null:{sample_df[col].isna().sum()}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(12, 5))

plt.subplot(3, 3, 1)
sns.histplot(sample_df['Price'], bins='auto', kde=True)
plt.title("Price DaÄŸÄ±lÄ±mÄ±")

plt.subplot(3, 3, 2)
sns.histplot(sample_df['Bedroom2'], bins='auto', kde=True)
plt.title("Bedroom DaÄŸÄ±lÄ±mÄ±")

plt.subplot(3, 3, 3)
sns.histplot(sample_df['Bathroom'], bins='auto', kde=True)
plt.title("Bathroom DaÄŸÄ±lÄ±mÄ±")

plt.subplot(3, 3, 4)
sns.histplot(sample_df['Car'], bins='auto', kde=True)
plt.title("Car DaÄŸÄ±lÄ±mÄ±")

plt.subplot(3, 3, 5)
sns.histplot(sample_df['Landsize'], bins='auto', kde=True)
plt.title("Landsize DaÄŸÄ±lÄ±mÄ±")

plt.subplot(3, 3, 6)
sns.histplot(sample_df['BuildingArea'], bins='auto', kde=True)
plt.title("BuildingArea DaÄŸÄ±lÄ±mÄ±")

plt.subplot(3, 3, 7)
sns.histplot(sample_df['YearBuilt'], bins='auto', kde=True)
plt.title("YearBuilt DaÄŸÄ±lÄ±mÄ±")

plt.subplot(3, 3, 8)
sns.histplot(sample_df['Lattitude'], bins='auto', kde=True)
plt.title("Lattitude DaÄŸÄ±lÄ±mÄ±")

plt.subplot(3, 3, 9)
sns.histplot(sample_df['Longtitude'], bins='auto', kde=True)
plt.title("Longtitude DaÄŸÄ±lÄ±mÄ±")

plt.tight_layout()
plt.grid()
plt.show()
```

```python
# null deÄŸerleri buna gÃ¶re doldurucam:
# price (int) -> mean
# bedroom (int) -> mode
# bathroom (int) -> mode
# car (int) -> mode
# landsize (int) -> median
# buildingarea (int) -> median
# yearbuild (date) -> median
# lattitude (float) -> mean
# longtitude (float) -> mean  
```

```python
sample_df['Price'] = sample_df['Price'].fillna(sample_df['Price'].mean()).round().astype('Int64')

sample_df['Bedroom2'] = sample_df['Bedroom2'].fillna(sample_df['Bedroom2'].mode()[0]).astype('Int64')

sample_df['Bathroom'] = sample_df['Bathroom'].fillna(sample_df['Bathroom'].mode()[0]).astype('Int64')

sample_df['Car'] = sample_df['Car'].fillna(sample_df['Car'].mode()[0]).astype('Int64')

sample_df['Landsize'] = sample_df['Landsize'].fillna(sample_df['Landsize'].median()).astype('Int64')

sample_df['BuildingArea'] = sample_df['BuildingArea'].fillna(sample_df['BuildingArea'].median()).round().astype('Int64')

sample_df['YearBuilt'] = sample_df['YearBuilt'].fillna(sample_df['YearBuilt'].median()).astype('Int64')

sample_df['Lattitude'] = sample_df.groupby('Suburb')['Lattitude'].transform(lambda x: x.fillna(x.mean()))
sample_df['Longtitude'] = sample_df.groupby('Suburb')['Longtitude'].transform(lambda x: x.fillna(x.mean()))

sample_df['Lattitude'] = sample_df['Lattitude'].fillna(sample_df['Lattitude'].mean())
sample_df['Longtitude'] = sample_df['Longtitude'].fillna(sample_df['Longtitude'].mean())

sample_df['Date'] = pd.to_datetime(sample_df['Date'], format='%d/%m/%Y')
sample_df['Postcode'] = sample_df['Postcode'].astype('Int64')
sample_df['Propertycount'] = sample_df['Propertycount'].astype('Int64')

print(sample_df.isna().sum())
print(sample_df.dtypes)
```
